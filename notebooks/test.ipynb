```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
imagePath = str(input())
```

```python
img = imageio.imread(imagePath)
```

```python
print(type(img))
print(img.shape)
print(type(img[0,0]))
print(img[0,0])
print(img[:5,:5])
```

```python
plt.imshow(img, cmap="gray")
plt.colorbar()
```

```python
img_inverted = np.zeros(img.shape, dtype=np.uint8)
print(type(img_inverted))
print(img_inverted.shape)
print(type(img[0,0]))
print(img_inverted[:5,:5])
print('Max value:', np.max(img))
print('Min value:', np.min(img))
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        img_inverted[row, col] = 255 - img[row, col]
```

```python
print(img_inverted[:5,:5])
print("Max value:", np.max(img_inverted))
print('Min value:', np.min(img_inverted))
```

```python
plt.imshow(img_inverted, cmap="gray")
plt.colorbar()
```

```python
imageio.imwrite('arara_inverted.jpg', img_inverted)
```
