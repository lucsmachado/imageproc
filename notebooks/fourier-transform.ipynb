```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
```

```python
x = np.arange(-1, 0.9, 0.01)
f = 5 + 2*x + 1*x**2
plt.figure(figsize=(12,3))
plt.plot(x, f, '.')
```

```python
sx = np.zeros(round(len(x)/20.0))
sf = np.zeros(round(len(x)/20.0))
print(sx, sf)
```

```python
j = 0
for i in range(0, len(x)+1, 21):
    sx[j] = x[i]
    sf[j] = f[i]
    j = j+1
    
print(sx)
print(sf)
```

```python
plt.figure(figsize=(12,3))
plt.plot(sx, sf, '.')
```

```python
N = 2
Y = sf
A = np.zeros([len(Y), N+1])
A[:,0] = sx**N
A[:,1] = sx**(N-1)
A[:,2] = 1

print(A)
print(Y)
```

```python
C1 = inv(np.matmul(np.transpose(A), A))
C2 = np.matmul(np.transpose(A), Y)
C = np.matmul(C1, C2)
print(C)
```

```python
t = 2
print('Sine: %.2f'% np.sin(t))
print('Cosine: %.2f'% np.cos(t))
```

```python
T = 2*np.pi
print('Sine: %.2f'% np.sin(t+T))
print('Cosine: %.2f'% np.cos(t+T))
```

```python
for i in range(6):
    print('Sine: %.2f'% np.sin(t+T*i))
    print('Cosine: %.2f'% np.cos(t+T*i))
```

```python
t = np.arange(0, T*5, 0.1)

fcos = np.cos(t)
fsin = np.sin(t)

plt.figure(figsize=(9,3))
plt.plot(t, fcos)
plt.plot(t, fsin)
```
