```python
import imageio.v2 as imageio
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 1, 0.001)
print('Number of observations: ', t.shape[0])
```

```python
my_sine = np.sin(2*np.pi*t)
plt.figure(figsize=(10,4))
plt.plot(t, my_sine)
```

```python
my_sine_4 = np.sin(2*np.pi*t*4)
plt.figure(figsize=(10,4))
plt.plot(t, my_sine_4)
```

```python
my_cos_4 = np.cos(2*np.pi*t*4)
plt.figure(figsize=(10,4))
plt.plot(t, my_cos_4)
```

```python
my_cos_18 = np.cos(2*np.pi*t*18)
plt.figure(figsize=(10,4))
plt.plot(t, my_cos_18)
```

```python
my_function = my_sine + my_sine_4 + my_cos_4 + my_cos_18
plt.figure(figsize=(10,4))
plt.plot(t, my_function)
```

```python
omega = 4
match_sin_4 = my_function * np.sin(omega*2*np.pi*t)
match_cos_4 = my_function * np.cos(omega*2*np.pi*t)
print('Sum of matching sines at 4Hz = %.2f' % np.sum(match_sin_4))
print('Sum of matching cosines at 4Hz = %.2f' % np.sum(match_cos_4))
```

```python
omega = 3
match_sin_3 = my_function * np.sin(omega*2*np.pi*t)
match_cos_3 = my_function * np.cos(omega*2*np.pi*t)
print('Sum of matching sines at 3Hz = %.2f' % np.sum(match_sin_3))
print('Sum of matching cosines at 3Hz = %.2f' % np.sum(match_cos_3))
```

```python
plt.figure(figsize=(10,4))
plt.plot(t, my_function, '-k')
plt.plot(t, match_sin_3, '--r')
plt.title('Matching sine at 3Hz')
```

```python
plt.figure(figsize=(10,4))
plt.plot(t, my_function, '-k')
plt.plot(t, match_sin_4, '--r')
plt.title('Matching sine at 4Hz')
```

```python
def match_freqs(f, t, max_freq):
    print('Coefficients of sine and cosine matching')
    print('omega\tsine\tcosine')
    
    for omega in np.arange(0, max_freq):
        match_sin = np.sum(f * np.sin(omega*2*np.pi*t))
        match_cos = np.sum(f * np.cos(omega*2*np.pi*t))
        print('%d\t%.1f\t%.1f' % (omega, match_sin, match_cos))
```

```python
match_freqs(my_function, t, 22)
```
